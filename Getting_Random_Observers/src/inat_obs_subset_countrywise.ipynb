````
%r

# old version of function to fetch observations for a given country, with a cutoff for active observers only
fetch_countrywise_old <- function(selected_country, cutoff = 20){

selected_bb <- filter(bounding_boxes,iso2c == selected_country)

inat_subset <-
  inat_open_data %>%
  filter(longitude > !!selected_bb$long_min) %>%
  filter(longitude < !!selected_bb$long_max) %>%
  filter(latitude > !!selected_bb$lat_min) %>%
  filter(latitude < !!selected_bb$lat_max)

# fetch coordinates to R env
longs <- pull(inat_subset,longitude)
lats <- pull(inat_subset,latitude)

# compare to shapefiles in local db, and push back to spark
sdf_copy_to(sc,tibble(country=maps::map.where(database = "world", x = longs, y = lats)),name = "countries", overwrite = TRUE)
  
# merge the country information to the observers, and filter so it's only from the chosen country
inat_subset <- sdf_bind_cols(inat_subset,sdf_sql(sc,"SELECT * FROM countries")) %>%
  filter(country == countrycode::countrycode(selected_country, origin= "iso2c", destination="country.name"))
# drop records of observers with less than the cutoff records
active_observers <- filter(count(inat_subset,observer_id),n>(cutoff-1)) %>% pull(observer_id)
inat_subset <- filter(inat_subset, observer_id %in% active_observers)
  
# print a short message about the result, the rowcount in the end is an expensive step that we don't really need
message(sprintf("passed %s: %s, subsetting resulted in %i records",selected_country, countrycode::countrycode(selected_country, origin = "iso2c", destination = "country.name"), sdf_nrow(inat_subset)))
# return subset
return(inat_subset)
  }
````

````
%r
# new version of the above function, this time implementing rounding in the coordinates so not every observation needs to be georeferenced, and also implementing an alternative filter mechanism so if there is no exact match to a country in the result list, the most common country gets chosen.

fetch_countrywise <- function(selected_country, cutoff = 20, decimals_to_round_to = 2){

selected_bb <- filter(bounding_boxes,iso2c == selected_country)

inat_subset <-
  inat_open_data %>%
  filter(longitude > !!selected_bb$long_min) %>%
  filter(longitude < !!selected_bb$long_max) %>%
  filter(latitude > !!selected_bb$lat_min) %>%
  filter(latitude < !!selected_bb$lat_max)

# check size of subset, for some countries this is very big thus requireing rounding of the coordinates so we have less places to reverse georeference
message(sprintf("The selected subset has %i rows",sparklyr::sdf_nrow(inat_subset)))


# Let's try to avoid having to reverse geocode every observation, as for some big countries R can't handle this out of the box (object size limits?)
# SEE ALSO: https://gis.stackexchange.com/questions/8650/measuring-accuracy-of-latitude-and-longitude/

inat_rounded <- 
  inat_subset %>%
    mutate(longitude_rounded = round(longitude,decimals_to_round_to),
           latitude_rounded = round(latitude,decimals_to_round_to)) %>%
    mutate(loc_id = paste0(longitude_rounded,latitude_rounded)) 

inat_locs <- inat_rounded %>%
  select(loc_id, longitude_rounded, latitude_rounded) %>%
  sparklyr::distinct()

# Check howmany unique locations we end up with after rounding
message(sprintf("Found %i unique localities after rounding",sparklyr::sdf_nrow(inat_locs)))

# reverse georeference
## use base:with instead of dplyr:pull to do reverse georeferencing in the local r environement

rev_georeferenced <- with(sparklyr::collect(inat_locs), 
       tibble(
         loc_id = loc_id,
         country = maps::map.where(database = "world", x = longitude_rounded, y = latitude_rounded)
       )
    )
sdf_copy_to(sc,rev_georeferenced,name = "cc_rounded", overwrite = TRUE)

# alternative, slower but base:with() warns about it's use in functions    
# sdf_copy_to(sc,
#            tibble(loc_id = pull(inat_locs,loc_id),
#                   country=maps::map.where(database = "world", x = pull(inat_locs,longitude_rounded), y = pull(inat_locs,latitude_rounded))),
#            name = "cc_rounded", overwrite = TRUE)

# join back to the original subset to attatch the localities to all of the observations
#inat_rounded_revgeoloc <- sparklyr::left_join(inat_rounded,sdf_sql(sc,"SELECT * FROM cc_rounded"), by = "loc_id")
inat_rounded_revgeoloc <- sparklyr::left_join(inat_rounded,spark_read_table(sc,"cc_rounded"), by = "loc_id")
sdf_register(inat_rounded_revgeoloc,name="countries")

# filter mechanism, drop observations that didn't geocode to the selected country
cc_tally <- sdf_sql(sc,"SELECT country,COUNT(*) as N FROM countries GROUP BY country ORDER BY N DESC") %>% sdf_collect()
if(countrycode::countrycode(selected_country, origin= "iso2c", destination="country.name") %>% stringr::str_replace("&","and") %in% pull(cc_tally,country)){
  inat_subset <- spark_read_table(sc,"countries") %>%
    filter(country == stringr::str_replace(countrycode::countrycode(selected_country, origin= "iso2c", destination="country.name"),"&","and"))} else {
  most_common_country <- pull(filter(cc_tally,N == max(N)),country)
  inat_subset <- spark_read_table(sc,"countries") %>%
    filter(country == !!most_common_country)
}

# drop records of observers with less than the cutoff records
active_observers <- filter(count(inat_subset,observer_id),n>(cutoff-1)) %>% pull(observer_id)
inat_subset <- filter(inat_subset, observer_id %in% active_observers)
  
# print a short message about the result, the rowcount in the end is an expensive step that we don't really need
message(
    sprintf("passed %s: %s, subsetting resulted in %i records",
    selected_country,
    countrycode::countrycode(selected_country, origin = "iso2c", destination = "country.name"), sdf_nrow(inat_subset)
        )
    )
# return subset
return(inat_subset)
  }
````

In a loop, so we have intermediate saving

````
%r
for(selected_cc in euro_iso2c[1:length(euro_iso2c)]){
  # we don't need to create a file we already have
  message(sprintf("considering %s", selected_cc))
  if(file.exists(sprintf("/dbfs/FileStore/%s_inat.rds",selected_cc))){
    message(sprintf("We already have %s_inat.rds for %s",
                    selected_cc,
                    countrycode::countrycode(selected_cc,
                                             origin = "iso2c",
                                             destination = "country.name")
                   )
           )
    next
  }
  
  out_sp <- fetch_countrywise(selected_cc) %>%
    sdf_collect(
    #  impl = "row-wise-iter"
    )
  # save the file, except if it's length zero. Then there is probably a problem with the translation between maps::map.where and countrycode::countrycode(destination="country.name")
  
  if(nrow(out_sp)>0){
    saveRDS(out_sp,
            file=sprintf("/dbfs/FileStore/%s_inat.rds",selected_cc)
           )
  message("saved, NEXT!!")} else { message(sprintf("problem with %s, not saved",selected_cc))}
}
````
